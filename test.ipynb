```python
import numpy as np

x = np.random.randint(0,256,1000000)

print(x)
print(np.bincount(x,minlength=256))
```

```python

def get_frequency(image:np.array,num_bins=None):

    image = np.ravel(image)
    bins_range = int(256/num_bins)
    freq = np.bincount(image,minlength=256)

    bin_freq = np.zeros(shape=num_bins)


    for i in range(num_bins-1):
        bin_freq[i] = np.sum(freq[i*bins_range:i*bins_range+bins_range]) # sliding window (window size = bins_range) step = bins_range
    
    i=num_bins-1
    bin_freq[-1]=np.sum(freq[i*bins_range:])
    return bin_freq

def get_frequency2(image: np.array, num_bins: int):
    # Flatten the image into a 1D array
    image = np.ravel(image)
    
    # Calculate the frequency of each pixel value (0-255)
    freq = np.bincount(image, minlength=256)
    
    # Calculate bin ranges
    bins_range = int(256 / num_bins)
    
    # Use np.add.reduceat to sum values for each bin
    indices = np.arange(0, 256, bins_range)
    bin_freq = np.add.reduceat(freq, indices)
    
    # If there are leftover values for the last bin (in case 256 is not divisible by num_bins)
    if len(bin_freq) > num_bins:
        # Merge the last part into the final bin
        bin_freq[-2] += bin_freq[-1]
        bin_freq = bin_freq[:-1]
    
    return bin_freq
```
